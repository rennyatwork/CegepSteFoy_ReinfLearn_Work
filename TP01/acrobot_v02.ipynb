```python
# 2. Import necessary libraries
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output  # For displaying images in Jupyter
from PIL import Image  # For image manipulation
```

```python
# 3. Initialize the Acrobot environment
env = gym.make('Acrobot-v1', render_mode='rgb_array')
```

```python
# 4. Create a function to choose a random action (for demonstration)
def choose_action(state):
    return env.action_space.sample()  # Random action for now
```

```python
# 5. Set up parameters
episodes = 1  # Number of episodes to run the agent
max_steps = 5  # Maximum steps per episode

# 6. Run the agent and visualize the environment step-by-step
for episode in range(episodes):
    state, info = env.reset()  # Reset the environment for each episode
    done = False
    step = 0
    
    while not done and step < max_steps:
        # Render the environment and capture the frame
        env.render()  # This creates a separate window for rendering

        # Choose an action using the function
        action = choose_action(state)

        # Apply the action to the environment
        next_state, reward, terminated, truncated, info = env.step(action)

        # Convert the rendering to an image for inline display
        img = Image.fromarray(env.render())  # Get the current frame as an image
        plt.figure(figsize=(5, 5))
        plt.imshow(img)  # Display the frame
        plt.axis('off')  # Turn off axis numbers
        display(plt.gcf())  # Display the current frame in Jupyter
        plt.close()  # Close the plot to avoid displaying the old frame again

        # Update step count and check if the episode is done
        step += 1
        done = terminated or truncated
        
        # Optional: slow down the display for visibility
        # time.sleep(0.1)  # Uncomment this for a pause between frames
    
    print(f"Episode {episode + 1} finished in {step} steps.")

# 7. Close the environment when done
env.close()
```
